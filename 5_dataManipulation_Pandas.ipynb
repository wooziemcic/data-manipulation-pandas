```python
import pandas as pd
```

### Creating a Series Object

```python
series = pd.Series([1,2,3,4,5], index = ['a','b','c','d','e'])
series
```

### Creating a Series Object through Dictionary

```python
# Note that the keys become the index values and the values will remain as values
series = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
pd.Series(series)
```

### Creating a Data Frame Object


```python
# Note that dataframe is multi-dimensional with label
dataframe = {'Name':['Aayush','Abhaya','Abhas'], 'Marks':[93,85,82]}
pd.DataFrame(dataframe, index=['i', 'ii', 'iii'])
```
